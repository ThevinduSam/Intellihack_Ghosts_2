# Dataset

```python
from google.colab import drive
import json

# Mount Google Drive
drive.mount('/content/drive')

# Assuming the .json file is in the root directory of your Google Drive
filename = '/content/drive/My Drive/ML_Task2/is_train.json'

# Open the .json file
with open(filename) as f:
    # Load the JSON data into a Python object
    data = json.load(f)

# Assuming the JSON file contains a list at the top level
# 'data' now holds the contents of the JSON file as a Python list
X_train = []
y_train = []
for i in range(len(data)):
    for j in range(2):
        if j == 0:
            X_train.append(data[i][j])
        else:
            y_train.append(data[i][j])

```

```python
# Assuming the .json file is in the root directory of your Google Drive
filename = '/content/drive/My Drive/ML_Task2/is_test.json'

# Open the .json file
with open(filename) as f:
    # Load the JSON data into a Python object
    tdata = json.load(f)

# Assuming the JSON file contains a list at the top level
# 'data' now holds the contents of the JSON file as a Python list
X_test = []
y_test = []
for i in range(len(tdata)):
  if i%4 == 0:
    for j in range(2):
        if j == 0:
            X_train.append(data[i][j])
        else:
            y_train.append(data[i][j])

  else:
    for j in range(2):
        if j == 0:
            X_test.append(data[i][j])
        else:
            y_test.append(data[i][j])

print(X_test)
print(y_test)
print(X_train)
print(y_train)
```

# Algorithm

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

class IntentClassifier:
    def __init__(self):
        self.pipeline = Pipeline([
            ('vectorizer', TfidfVectorizer(ngram_range=(1, 2))),  # Convert text to TF-IDF representation with bi-grams
            ('classifier', MultinomialNB())                     # Naive Bayes classifier
        ])

    def train(self, X_train, y_train):
        self.pipeline.fit(X_train, y_train)

    def predict(self, X_test):
        return self.pipeline.predict(X_test)
```

# Test set predictions and accuracy

```python
if __name__ == "__main__":
    # Example training data


    # Initialize and train the classifier
    classifier = IntentClassifier()
    classifier.train(X_train, y_train)

    # Example test data


    # Predict intents for test data
    predictions = classifier.predict(X_test)

    # Output predictions
    for text, intent in zip(X_test, predictions):
        print(f"Text: {text} --> Predicted Intent: {intent}")
```

```python
from sklearn.metrics import accuracy_score

for text, intent, true_intent in zip(X_test, predictions, y_test):
        print(f"Text: {text} --> Predicted Intent: {intent}, True Intent: {true_intent}")

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy}")
```

# Training set predictions and accuracy

```python
if __name__ == "__main__":
    # Example training data


    # Initialize and train the classifier
    classifier = IntentClassifier()
    classifier.train(X_train, y_train)

    # Example test data


    # Predict intents for test data
    predi = classifier.predict(X_train)

    # Output predictions
    for text, intent in zip(X_train, predi):
        print(f"Text: {text} --> Predicted Intent: {intent}")
```
